Tarea de Gráfica de las funciones implícitas

```python
from sympy import symbols, Eq
from sympy import plot_implicit

x, y = symbols('x y')
eq = Eq(x**2 + y**3-2*y, 3)
print(eq)
plot_implicit(Eq(x**2 + y**3-2*y,3), (x,-4,4),(y,-4,3));
```

Actividad de funciones paramétricas

```python
from sympy import symbols, cos, sin, exp, pi
from sympy import plot_parametric
t = symbols('t')
def x(t):
  return cos(t)+1/cos(7*t)+1/3*sin(17*t)
def y(t):
  return sin(t)+1/2*sin(7*t)+1/3*cos(17*t)

plot_parametric(x(t), y(t), (t, 0, 14*pi));
plot_parametric(x(t), y(t), (t, 0, 14*pi));
```

Actividad de coordenadas polares

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 4000)

#Aqué tengo una duda sobre como poner la a, para que me aparezcan la familia de funciones, por eso aqui lo hice con a=6 y a=10

r = 6 *(1+ np.cos(theta))
x = r * np.sin(theta)
y = r * np.cos(theta)

plt.axis('off');
plt.axis('equal');
plt.plot(x,y);


r1 = 10 *(1+ np.cos(theta))
x = r1 * np.sin(theta)
y = r1 * np.cos(theta)

plt.axis('off');
plt.axis('equal');
plt.plot(x,y);
```
