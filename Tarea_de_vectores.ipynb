Mostrar que se cumple la propiedad conmutativa con la operaci√≥n suma de vectores

```python
import matplotlib.pyplot as plt

def sumaComplejos(z1,z2): # as√≠ se definen las funciones
  return z1[0]+z2[0],z1[1]+z2[1]


def sumaVectores(z1,z2):
  return z1[0]+z2[0],z1[1]+z2[1]

z1=(2,1)
z2=(3,2)

suma=sumaComplejos(z1,z2)

#grafica la suma

plt.quiver(0,0,z1[0],z1[1], color="g", units="xy", scale=1)
plt.quiver(z1[0], z1[1], z2[0], z2[1], color="b", units="xy", scale=1)
plt.quiver(0, 0, suma[0], suma[1], color="r", units="xy", scale=1)

plt.quiver(0,0,z2[0],z2[1], color="b", units="xy", scale=1)
plt.quiver(z2[0], z2[1], z1[0], z1[1], color="g", units="xy", scale=1)



plt.axis("scaled") #Esto es para que los vectores no se muevan y se ajuste la maquina
plt.xlim(0,5) # esto te dice de donde a donde en el eje x
plt.ylim(0,5) # esto te dice de donde a donde en el eje y
plt.show()
```

Mostrar que se cumple la propiedad asociativa con la operaci√≥n suma de vectores

```python
import matplotlib.pyplot as plt

def sumaComplejos(z1,z2): # as√≠ se definen las funciones
  return z1[0]+z2[0],z1[1]+z2[1]


def sumaVectores(z1,z2):
  return z1[0]+z2[0],z1[1]+z2[1]

z1=(2,1)
z2=(3,2)
z3=(4,3)

suma=sumaComplejos(z1,z2)

#grafica la suma

plt.quiver(0,0,z1[0],z1[1], color="g", units="xy", scale=1)
plt.quiver(z1[0], z1[1], z2[0], z2[1], color="b", units="xy", scale=1)
plt.quiver(0, 0, suma[0], suma[1], color="r", units="xy", scale=1)

plt.quiver(suma[0],suma[1], z3[0],z3[1], color="y", units="xy", scale=1)
plt.quiver(0, 0, suma[0], suma[1], color="r", units="xy", scale=1)

suma1=sumaComplejos(z2,z3)

plt.quiver(0,0,z2[0],z2[1], color="b", units="xy", scale=1)
plt.quiver(z2[0], z2[1], z3[0], z3[1], color="y", units="xy", scale=1)
plt.quiver(0, 0, suma1[0], suma1[1], color="r", units="xy", scale=1)
plt.quiver(suma1[0],suma1[1], z1[0],z1[1], color="g", units="xy", scale=1)







plt.axis("scaled") #Esto es para que los vectores no se muevan y se ajuste la maquina
plt.xlim(0,9) # esto te dice de donde a donde en el eje x
plt.ylim(0,7) # esto te dice de donde a donde en el eje y
plt.show()
```
